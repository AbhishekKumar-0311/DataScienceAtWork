# Basic Analysis

### value_counts()

```python
have.Edition.value_counts()
```

```python
have.City.value_counts()
```

```python
# The output is sorted on count(Gender)
# By default dropna=True : Don't include counts of NaN

have.Gender.value_counts(ascending=True)
```

### sort_values()

```python
# Outputs a series sorted ascending, by default

have.City.sort_values()
```

```python
# Outputs a dF sorted ascending, by default

have.sort_values(by='City')
```

```python
# Outputs a dF sorted based on multiple SeriesNames or ColumnNames

have.sort_values(by=['Edition','Gender','Athlete'],ascending=False)
```

### Boolean indexing

```python
have.Sport == 'Tennis'
```

```python
have[have.Sport == 'Tennis']
```

```python
have[(have.Medal == 'Gold') & (have.Gender == 'Women')]
```
