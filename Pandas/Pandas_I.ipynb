# Groupby

```python
import pandas as pd

have = pd.read_csv('E:\VCS\GitHub\DataScienceAtWork\data\olympics.csv', skiprows=4)
have.head()
```

Note:
1. Groupby perform 3 important functions.
2. Split a dataframe into groups based on some criteria.
3. Apply a function to each group independently.
4. Combine the results into a dataframe.
5. Groupby object is not a dataframe but a group of dataframes.

### The Groupby object

```python
have.groupby('Edition')
```

```python
type(have.groupby('Edition'))
```

```python
list(have.groupby('Edition'))
```

## Iterating through groups

```python
for group_key, group_value in have.groupby('Edition'):
    print(group_key)
    print(group_value)
```

```python
type(group_value)
```

## Groupby computations

```python
have.groupby('Edition').size()
```

### agg([...])

```python
have.groupby('Edition').agg('count')
```

```python
have.groupby(['Edition','NOC','Medal']).agg(['min','max','count'])
```

```python
have.groupby(['Edition','NOC','Medal']).size()
```

### agg({  : [ ... ]})

```python
have.groupby(['Edition','NOC','Medal']).agg({'Edition':['min','max','count']})
```

# stack() and unstack()

```python
import pandas as pd

have = pd.read_csv('E:\VCS\GitHub\DataScienceAtWork\data\olympics.csv', skiprows=4)
have.head()
```

## Athletes winning medals in Beijing Olympics 100m or 200m track event

```python
mw = have[(have.Edition == 2008) & ( (have.Event == '100m') | (have.Event == '200m'))]
mw
```

```python
g = mw.groupby(['NOC','Gender','Discipline','Event']).size()
g
```

```python
df = g.unstack(['Discipline','Event'])
df
```

## stack()

```python
df
```

```python
df.stack()
```

Note:
1. Stack() return a dataFrame or a series.

```python
df.stack('Event')
```

```python
df.stack('Discipline')
```

```python
df.stack(['Discipline','Event'])
```

## unstack()

```python
df
```

```python
df.unstack()
```

```python
df.unstack('Gender')
```
