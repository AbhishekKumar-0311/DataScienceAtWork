## Plot types

### What were the different sports in the first olympics? Plot them using different graphs.

```python
# Creating a dF of 1st edition of Olympics

ed_one = have[have.Edition == 1896]
```

```python
# Plotting the different sports in 1st Olympics.
# The default chart with plot() is Line-chart.
# Insert semi-colon to remove the axes info.

ed_one.Sport.value_counts().plot(kind='line')
```

```python
# Plotting the Bar-chart different sports in 1st Olympics.
# The default chart with plot() is Line-chart.
# Insert semi-colon to remove the axes info.

ed_one.Sport.value_counts().plot(kind='bar');
```

```python
# Plotting the Horizontal Bar-chart different sports in 1st Olympics.
# The default chart with plot() is Line-chart.
# Insert semi-colon to remove the axes info.

ed_one.Sport.value_counts().plot(kind='barh');
```

```python
# Plotting the Pie-chart different sports in 1st Olympics.
# The default chart with plot() is Line-chart.
# Insert semi-colon to remove the axes info.

ed_one.Sport.value_counts().plot(kind='pie');
```

## Plot colors

```python
# Plotting the Pie-chart different sports in 1st Olympics.
# The default chart with plot() is Line-chart.
# Insert semi-colon to remove the axes info.

ed_one.Sport.value_counts().plot(kind='bar', color= 'green');
```

## Figure Size - figsize()

```python
# Plotting the different sports in 1st Olympics.
# The default chart with plot() is Line-chart.
# Insert semi-colon to remove the axes info.
# figsize=(width,height)

ed_one.Sport.value_counts().plot(kind='line',figsize=(10,3));
```

## Colormaps

```python
# 3 classes of colormaps: sequential diverging qualitative 
# sequential to represent that has ordering.
# diverging - when plotting represented deviate around middle vale,
# qualitative - is used to represent information that does not have ordering
# ed_one.Sport.value_counts().plot(kind='pie',color='Paired'); color= throws Error

ed_one.Sport.value_counts().plot(kind='pie',colormap='Paired');
```

# Seaborn basic plotting

```python
import seaborn as sbn
```

### How many medals have been won by men and women in the history of the Olympics. How many gold, silver and bronze medals were won for each gender?

```python
sbn.countplot(x='Medal', data=have, hue='Gender');
```

### Plot the number of medals won by the chinese team(men n women) in Beijing 2008 Olympics

```python
chn = have[(have.NOC=='CHN') & (have.Edition==2008)]
```

```python
chn.Gender.value_counts().plot(kind='pie',colormap='Paired');
```

```python
chn.Gender.value_counts().plot(kind='bar',colormap='Paired');
```

```python
import seaborn as sbn
sbn.countplot(x='Gender', data=have,palette='bwr');
```

### Plot the number of gold, silver and bronze for each gender

```python
import seaborn as sbn
```

```python
sbn.countplot(x='Medal', data=have, hue='Gender',palette='bwr', order=['Gold','Silver','Bronze']);
```
